```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Daten laden
df = pd.read_csv("cardio_train.csv")

# Optional: schnelle Sichtprüfung
print(df.shape)
print(df.columns.tolist())
df = df.dropna()

# WICHTIG: age ist in Tagen -> in Jahre umrechnen
df["age_years"] = (df["age"] / 365.25).round(1)

# 2) Histogramm: Altersverteilung
plt.figure()
df["age_years"].plot(kind="hist", bins=30)
plt.title("Altersverteilung (Jahre)")
plt.xlabel("Alter (Jahre)")
plt.ylabel("Anzahl")
plt.tight_layout()
plt.show()

# 3) Kreisdiagramm: Anteil mit/ohne CVD (Spalte 'cardio' 0/1)
counts = df["cardio"].value_counts().sort_index()
labels = ["Kein CVD (0)", "CVD (1)"] if 0 in counts.index and 1 in counts.index else counts.index.astype(str)

plt.figure()
plt.pie(counts, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Anteil kardiovaskuläre Erkrankung")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Richtig einlesen (Semikolon als Trennzeichen!)
df = pd.read_csv("cardio_train.csv", sep=";")

# 2) Alter (Tage -> Jahre)
df["age_years"] = (df["age"] / 365.25).round(1)

# 3) Histogramm: Altersverteilung
plt.figure()
df["age_years"].plot(kind="hist", bins=30)
plt.title("Altersverteilung (Jahre)")
plt.xlabel("Alter (Jahre)")
plt.ylabel("Anzahl")
plt.tight_layout()
plt.show()

# 4) Kreisdiagramm: Anteil mit/ohne CVD
counts = df["cardio"].value_counts().sort_index()
labels = ["Kein CVD (0)", "CVD (1)"] if set(counts.index) == {0, 1} else counts.index.astype(str)

plt.figure()
plt.pie(counts, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Anteil kardiovaskuläre Erkrankung")
plt.tight_layout()
plt.show()
```
